#### 3. Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
#NOVA TENTATIVA
#we want movie name, year, director and stars)
response = requests.get(url)
html = response.text
````

````python
#to look for the TITLES:
soup = BeautifulSoup(html, 'html.parser') #we used the default parser (html)
movietags = soup.select('td.titleColumn')
stars = soup.find_all('td', attrs = {'class': 'ratingColumn imdbRating'})

#all years:
clean_years_list = []
for i in range(0, 250):
    clean_years = movietags[i].find('span').get_text().replace('(','').replace(')','')
    #print(clean_years)
    clean_years_list.append(clean_years)
#print(clean_years_list)

#movie titles:
clean_title_list = []
for j in range(0, 250):
    clean_titles = movietags[j].find('a').get_text()
    clean_title_list.append(clean_titles)
#print(clean_title_list)

#directors:
clean_director_list = []
for k in range(0, 250):
    clean_directors = movietags[k].find('a').get('title').split(' (')[0]
    clean_director_list.append(clean_directors)  
#print(clean_director_list)

#stars:
clean_stars_list = []
for l in range(0, 250):
    clean_stars = stars[l].get_text(strip = True)
    clean_stars_list.append(clean_stars)
#print(clean_stars_list)

'''#movie titles:
movietags = soup.select('td.titleColumn') #selects all elements with the name titlecolumn
movietags[0].find('a').get_text() #clean movie title

#year:
movietags[0].find('span').get_text().replace('(','').replace(')','')

#Director:
movietags[0].find('a').get('title').split(' (')[0]

#stars:
stars = soup.find_all('td', attrs = {'class': 'ratingColumn imdbRating'})
stars[0].get_text(strip = True)'''
````

````python
pd_films = pd.DataFrame(clean_title_list, columns = ['Movie name'])
pd_year = pd.DataFrame(clean_years_list, columns = ['Initial release'])
pd_director = pd.DataFrame(clean_director_list, columns = ['Director name'])
pd_stars = pd.DataFrame(clean_stars_list, columns = ['Stars'])
final_dataframe = pd.concat([pd_films, pd_year, pd_director, pd_stars], axis = 1)
print(final_dataframe)
````

#### 3.1. Display the movie name, year and a brief summary of the top 10 random movies (IMDB) as a pandas dataframe.

````python
#This is the url you will scrape in this exercise
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
html = response.text
````

````python
# movie name, year and a brief summary
#
soup = BeautifulSoup(html, 'html.parser') #we used the default parser (html)
movietags = soup.select('td.titleColumn')
summary_tags = soup.find_all('td', attrs = {'class': 'posterColumn'})

#all years:
clean_years_list1 = []
for i in range(0, 11):
    clean_years = movietags[i].find('span').get_text().replace('(','').replace(')','')
    #print(clean_years)
    clean_years_list1.append(clean_years)
#print(clean_years_list1)

#movie titles:
clean_title_list1 = []
for j in range(0, 11):
    clean_titles = movietags[j].find('a').get_text()
    clean_title_list1.append(clean_titles)
#print(clean_title_list1)

#clean summary
clean_summary_list = []
for i in range(0, 11):
    url_summary = 'https://www.imdb.com' + str(summary_tags[i].find('a').get('href'))
    response_summary = requests.get(url_summary)
    soup = BeautifulSoup(response_summary.content, features = 'lxml')
    indiv_summary = soup.find_all('div', attrs = {'class': 'summary_text'})
    indiv_summary = indiv_summary[0].get_text(strip = True)
    clean_summary_list.append(indiv_summary)
#print(clean_summary_list)
````

````python
pd_films1 = pd.DataFrame(clean_title_list1, columns = ['Movie name'])
pd_year1 = pd.DataFrame(clean_years_list1, columns = ['Year'])
pd_summary1 = pd.DataFrame(clean_summary_list, columns = ['Summary'])

final_dataframe1 = pd.concat([pd_films1, pd_year1, pd_summary1], axis = 1)
print(final_dataframe1)
````
